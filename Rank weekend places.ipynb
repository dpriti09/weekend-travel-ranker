```python
import pandas as pd
```

```python
place = pd.read_csv('Top Indian Places to Visit.csv')
```

```python
(place.head())
```

```python
place.shape
```

```python
def rank_weekend_places(city_name):
   
    city_data = place[place['City'].str.contains(city_name, case=False, na=False)]
    city_data_sorted = city_data.sort_values(by='Google review rating', ascending=False)
    return city_data_sorted[['Name', 'Type', 'Google review rating', 'Entrance Fee in INR', 'time needed to visit in hrs', 'Best Time to visit']]
```

```python
city_name = "Kolkata"
ranked_places = rank_weekend_places(city_name)


(ranked_places)
```
